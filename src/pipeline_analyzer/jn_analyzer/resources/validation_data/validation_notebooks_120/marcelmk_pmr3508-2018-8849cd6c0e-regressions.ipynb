```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
train = pd.read_csv("../input/train.csv", index_col = 'Id')
```

```python
train
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(38)
```

```python
knn_score = cross_val_score(knn, train.drop('median_house_value', axis = 1), train['median_house_value'], cv = 10, scoring = 'r2')
```

```python
knn_score.mean()
```

Optei por utilizar o regressor LassoCV, uma vez que ele usa um cross-validation para determinar o parâmetro alpha

```python
from sklearn.linear_model import LassoCV
```

```python
lasso = LassoCV(cv = 10)
```

```python
lasso_score = cross_val_score(lasso, train.drop('median_house_value', axis = 1), train['median_house_value'], cv = 10, scoring = 'r2')
```

```python
lasso_score.mean()
```

Novamente optei pela versão do regressor que já se utiliza de cross-validation para determinar seu hiperparâmetro

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge = RidgeCV(cv = 10)
```

```python
ridge_score = cross_val_score(ridge, train.drop('median_house_value', axis = 1), train['median_house_value'], cv = 10, scoring = 'r2')
```

```python
ridge_score.mean()
```

Analisando as métricas utilizadas, chegamos que nesse caso, a regressão RIDGE obteve melhor desempenho que as demais
