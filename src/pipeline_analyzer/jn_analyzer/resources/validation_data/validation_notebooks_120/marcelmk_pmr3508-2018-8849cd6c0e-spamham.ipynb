```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../input/featdataset/train_data.csv", index_col = 'Id')
train.head()
```

```python
plt.figure(figsize=(15,10))
(train[train['ham'] == True].mean() - train[train['ham'] == False].mean())[train.columns[:54]].plot(kind = 'bar')
```

<p>Pelo plot anterior, podemos ter uma ideia de quais palavras apareceram com mais frequência em spams e quais apareceram com mais frequência em hams. Os valores negativos indicam uma maior ocorrência em spams, e positivos indicam maior ocorrência em hams.  </p>

```python
(train[train['ham'] == True].mean() - train[train['ham'] == False].mean())[train.columns[54:57]].plot(kind = 'bar')
```

```python
import sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
Xtrain = train[train.columns[0:57]]
Xtrain.head()
```

```python
Ytrain = train['ham']
Ytrain.head()
```

```python
NB = GaussianNB()
```

```python
scores = cross_val_score(NB, Xtrain,Ytrain, cv=10)
scores
```

```python
test = pd.read_csv('../input/featdataset/test_features.csv', index_col = 'Id')
test.head()
```

```python
NB.fit(Xtrain,Ytrain)
Ytest = NB.predict(test)
```

```python
pred = pd.DataFrame(index = test.index)
pred['ham'] = Ytest
pred.to_csv('submission.csv',index = True)
```
