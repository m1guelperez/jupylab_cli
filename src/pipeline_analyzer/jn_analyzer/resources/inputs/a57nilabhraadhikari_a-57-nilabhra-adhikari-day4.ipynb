```python
from matplotlib import pyplot as plt
x=['O+','A+','B+','AB+','O-','A-','B-','AB-']
y=[9,12,2,10,7,1,4,5]
plt.plot(x,y,'r--',x,y,'g^')
plt.title('Blood group distribution of 50 patients')
plt.ylabel('No of Patients')
plt.xlabel('Blood group')
plt.show()
```
