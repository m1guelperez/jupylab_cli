**#Data Wrangling**

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop('Cabin', axis=1, inplace=True)
dataset.head()
```

```python
sns.heatmap(dataset.isnull())
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull())
```

```python
dataset.head(3)
```

```python
sex = pd.get_dummies(dataset['Sex'], drop_first=True)
sex.head()
```

```python
embarked = pd.get_dummies(dataset['Embarked'], drop_first=True)
embarked.head(5)
```

```python
pcl = pd.get_dummies(dataset['Pclass'], drop_first=True)
pcl.head()
```

```python
dataset = pd.concat([dataset, sex, embarked, pcl], axis=1)
```

```python
dataset.head()
```

```python
dataset.drop(['Sex','Embarked','PassengerId','Name','Ticket'], axis=1, inplace=True)
```

```python
dataset
```

```python
dataset.drop(['Pclass'], axis=1, inplace=True)
```

```python
dataset.head(5)
```

**#Train Data**

```python
X = dataset.drop('Survived', axis=1)
y = dataset['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
prediction = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, prediction)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, prediction)
```
