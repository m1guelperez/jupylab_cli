```python
from math import sqrt

def pyth(cote1,cote2):
    rep=input("Vous désirez calculer la longueur de l'hypoténuse ?")
    if rep=='oui':
        longueur=sqrt(cote1**2+cote2**2)
    else :
        if cote1<cote2:
            longueur=sqrt(cote2**2-cote1**2)
        else :
            longueur=sqrt(cote1**2-cote2**2)
    return longueur
```
